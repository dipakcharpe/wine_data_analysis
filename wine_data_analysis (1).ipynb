```python
#using groupBy to find avg alcohol percentage qualitywise
quality_wise_alcohol = df_red.groupby(df_red.quality)['alcohol'].mean().reset_index().rename(columns = {'alcohol':'avg_alcohol_conc'})
quality_wise_alcohol
```

```python
#graphical visualysation of quality and acohol percentage
sns.scatterplot(data = quality_wise_alcohol, x='quality',y='avg_alcohol_conc')
plt.show()
```

```python
#looking for the correlation between alcohol and pH getting a best fitted line

print("The Rho is {}".format(np.corrcoef(df_red.alcohol, df_red.pH)[0][1]))

g = sns.jointplot(data = df_red, x = "alcohol", y = "pH", kind='reg', color='royalblue')
# ax.annotate(stats.pearsonr)
r, p = stats.pearsonr(df_red.alcohol, df_red.pH)
g.ax_joint.annotate(f'$\\rho = {r:.3f}, p = {p:.3f}$',
                    xy=(0.1, 0.9), xycoords='axes fraction',
                    ha='left', va='center',
                    bbox={'boxstyle': 'round', 'fc': 'powderblue', 'ec': 'navy'})
g.ax_joint.scatter(df_red.alcohol, df_red.pH)
g.set_axis_labels(xlabel="alcohol", ylabel="pH", size=15)
plt.tight_layout()
plt.show()
```

```python
#changing the datatype of quality columns from int64 to categorical
df_red['quality'] = pd.Categorical(df_red.quality)
df_red.dtypes
```

```python
#changing the datatype of quality columns from int64 to categorical
df_white['quality'] = pd.Categorical(df_white.quality)
df_white.dtypes
```

```python
# creating new cnditional column in dataset without using lamda function
df_red['quality_label'] = ['low' if x<=5 else ('medium' if x>5 and x<=7 else 'high' ) for x in df_red['quality']]
df_red
#df_red['wine category']= df_red['quality'].apply(lambda x :'low' if x <= 5 else ('medium' if x>5 and x<=7 else 'high' ))
```

```python
# creating new cnditional column in dataset using lambda function
df_white['quality_label'] = df_white.quality.apply(lambda x: 'low'if x<=5 else('medium' if x>5 and x<=7 else "high"))
df_white
```

```python
#counting wine category so that we could get the quality wise differance
quality_wise_category = df_red.groupby(df_red.quality_label)['wine_category'].count().reset_index().rename(columns = {'wine_category':'category_count'})
quality_wise_category
```

```python
#counting wine category so that we could get the quality wise differance
quality_wise_category = df_white.groupby(df_white.quality_label)['wine_category'].count().reset_index().rename(columns = {'wine_category':'category_count'})
quality_wise_category
```

```python
#concatinating redwine and white wine data
df = pd.concat([df_red ,df_white],axis =0)
df
```

```python
#shuffling the columns for further use
df = df.sample(frac = 1)
df
```
